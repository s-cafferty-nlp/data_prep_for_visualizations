```python
import pandas as pd
```

```python
with open('../data/proust.txt', 'r') as file:
     data = file.read()
len(data)
```

```python
split_data = data.split('\n')
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
vectorizer = CountVectorizer()
X = vectorizer.fit_transform([data])
```

```python
import pandas as pd

counts = X.toarray()
vocab = vectorizer.get_feature_names_out()
```

```python
X
```

```python
proust_count_df = pd.DataFrame(counts, columns=vocab).transpose().reset_index().rename(columns={'index':'word',0:'count'})
proust_count_df 
```

```python
proust_pos = pd.read_csv('../data/proust_pos.csv').rename(columns={'Unnamed: 0':'word','0':'pos'})
```

```python
proust_pos
```

```python
proust_pos_dict = {}

for x in range(len(proust_pos)):
    proust_pos_dict[proust_pos['word'].iloc[x].lower()] = proust_pos['pos'].iloc[x]

proust_pos_dict
```

```python
proust_count_df['pos'] = [proust_pos_dict[x] if x in set(proust_pos_dict.keys()) else 'NONE' for x in list(proust_count_df['word'])]
```

```python
proust_count_df[(proust_count_df['pos']=='ADJ') & (proust_count_df['count']>10)]
```

```python
proust_count_df[(proust_count_df['pos']=='VERB') & (proust_count_df['count']>10)  & ((proust_count_df['word'].str.slice(-2)=='er') | (proust_count_df['word'].str.slice(-2)=='ir') | (proust_count_df['word'].str.slice(-2)=='re'))]
```
